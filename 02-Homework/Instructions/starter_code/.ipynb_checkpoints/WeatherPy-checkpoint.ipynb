## Generate Cities List

```python
# Genertate Cities List : 643
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime
from scipy.stats import linregress
import scipy.stats as st

# Import API key

from api_keys import weather_api_key
print(weather_api_key)
# Incorporated citipy to determine city based on latitude and longitude

from citipy import citipy

# Output File (CSV)

# output_data_file = "cities.csv"
# Range of latitudes and longitudes

lat_range = (-90, 90)
lng_range = (-180, 180)

# create random latitudes and Longitude: change size to only 25 for testing
random_lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
random_longs = np.random.uniform(lng_range[0], lng_range[1], size=1500)

# zip the list together
zipped_lat_longs = zip(random_lats, random_longs)
lats_longs_list = list(zipped_lat_longs)
```

```python
# for testing generate one city name
lat_long = lats_longs_list[0]
citipy.nearest_city(lat_long[0], lat_long[1]).city_name
# test_city = citipy.nearest_city(lat_long[0], lat_long[1]).city_name
# print(test_city)
```

```python
cities = []
for lat_long in lats_longs_list:
        city_name = citipy.nearest_city(lat_long[0], lat_long[1]).city_name
        
        if city_name not in cities:
            cities.append(city_name)
        
    
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
print(query_url)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# List for holding reponse information
lat = []
temp = []
temp_max = []
humidity = []
wind_speed = []
lon = []
date = []
country = []
cloudiness = []
citys = []

# Loop through the list of cities and request for data on each
print("Beginning Data Retrieval")
print("--------------------------------------------")
count = 0
set = 1
for idx, city in enumerate(cities):
#     count = count + 1
#     # To avoid api call rate limits, get city weather data for 50 cities,
#     # then sleep for 5 seconds, and then continue with next 50 cities and so on...
#     if count == 51:
    count = 1
    set = set + 1
#         time.sleep(5)
#         print(f"Processing Record {count} of Set {set} | {city}")
    try:
        print(f"Processing Record {count} of Set {set} | {city}")
        response = requests.get(query_url + city).json()
#         print(response)
        lat.append(response['coord']['lat'])
        lon.append(response['coord']['lon'])
        temp.append(response['main']['temp'])
        temp_max.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        wind_speed.append(response['wind']['speed'])
        date.append(response['dt'])
        country.append(response['sys']['country'])
        cloudiness.append(response['clouds']['all'])
        citys.append(city)
    except KeyError:
#         print("City not found. Skipping...")
        lat.append(np.nan)
        lon.append(np.nan)
        temp.append(np.nan)
        temp_max.append(np.nan)
        humidity.append(np.nan)
        wind_speed.append(np.nan)
        date.append(np.nan)
        country.append(np.nan)
        cloudiness.append(np.nan)
        citys.append(np.nan)
print("------------------------------------------------")
print("Data Retrieval Complete")
print("------------------------------------------------")
```

```python
print(citys)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
Weather_city_df = pd.DataFrame({'City Name':citys,
                                'Country Name':country,
                                'Date':date,
                                'Latitude':lat,
                                'Longitude':lon,
                                'City Temp':temp,
                                'City Temp Max':temp_max,
                                'City Humidity':humidity,
                                'City Wind Speed':wind_speed,
                                'City Clouds':cloudiness})

# Drop any cities that were skipped because they could not be found using the OpenWeatherMap API.
Weather_city_df = Weather_city_df.dropna(how="any")
print(Weather_city_df.count())
Weather_city_df
```

```python
# save the DataFrame as csv
Weather_city_df.to_csv("Weather_city_data.csv", index = False)
```

## Linear Regression

```python
# create Northern and Southern Hemisphere DF
north_hem = Weather_city_df.loc[Weather_city_df["Latitude"] >= 0]
south_hem = Weather_city_df.loc[Weather_city_df["Latitude"] < 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Add the Linear regression for Northern Hemishpere
x_values = north_hem["Latitude"]
y_values = north_hem["City Temp Max"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values, facecolor = "steelblue", edgecolor = "black")
plt.plot(x_values,regress_values, "r-")
plt.annotate(line_eq, (6,10), fontsize=15,color="red")
# Set a title
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
# Print r value
print(f"The r-value is: {rvalue**2}")
print(line_eq)
plt.show()
# save the plotted figure as .pngs
plt.savefig("Northern Hemisphere - Max Temp vs. Latitude Linear Regression.png")
```
